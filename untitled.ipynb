```python
import pandas as pd
import duckdb
```

```python
conn = duckdb.connect('database/database.duckdb')
```

## First method pass the SQL "interactively 

```python
# Run query - VIEW 
conn.sql("SELECT * FROM Site LIMIT 5")
```

```python
# Get the data from query 
conn.sql("SELECT * FROM Site LIMIT 5").fetchall()
```

```python
# Return one row 
conn.sql("SELECT * FROM Site LIMIT 5").fetchone()
```

```python
# Get data in dataframe 
conn.sql("SELECT * FROM Site LIMIT 5").df()
```

## Using Curser

```python
# Create a curser 
cur = conn.cursor()
```

```python
# curser object - create connection for specific SQL query 
cur.execute("SELECT * FROM Site")
```

```python
# you can apply the fetchall()... 
cur.fetchall()
```

```python
# Getting on row 
cur.execute("SELECT * FROM Site")
cur.fetchone()
```

```python
cur.fetchone()[0]
```

```python
# we want to iterate throughout the rows and print the site code 
cur.execute("SELECT * FROM Site")

for row in cur.fetchall(): 
    print(row[0])
```

```python
cur.execute("SELECT * FROM Site")

while True: # create an infinite loop
    row = cur.fetchone() 
    if row == None: # If nothing in row, break
        break
    print(f"Site code: {row[0]}")
```

```python
cur.execute("""CREATE TEMP TABLE temp_nests AS 
            SELECT * FROM Bird_nests""")
```

```python
cur.execute("SELECT * FROM temp_nests LIMIT 5").fetchall()
```
